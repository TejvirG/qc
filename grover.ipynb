```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(2)

qc.h([0,1])

qc.cz(0,1)

qc.h([0,1])
qc.x([0,1])
qc.h(1)
qc.cx(0,1)
qc.h(1)
qc.x([0,1])
qc.h([0,1])

print(qc.draw())

sv = Statevector.from_instruction(qc)
print("Probabilities:", sv.probabilities_dict())
```
